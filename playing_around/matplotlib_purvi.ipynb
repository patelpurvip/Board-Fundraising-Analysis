```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame({"Q1":[10,16,20,35],"Q2":[20,3,18,30],"Q3":[5,15,0,15],"Q4":[2,15,1,20]},index=[2020,2019,2018,2017])
df
```

```python
list_2020 = df.iloc[0]
list_2019 = df.iloc[1]
list_2018 = df.iloc[2]
list_2017 = df.iloc[3]

list_Qs = ['Q1','Q2','Q3','Q4']
x = [1,2,3,4]
```

```python
fig = plt.figure(figsize=(10,6))

plt.plot(x, list_2020, label = "2020")
plt.plot(x, list_2019, label = "2019")
plt.plot(x, list_2018, label = "2018")
plt.plot(x, list_2017, label = "2017")

plt.title("Sales per Quarter")
plt.xlabel("Quarters")
plt.ylabel("Sales")
plt.legend()
plt.xticks(x, list_Qs, rotation=45)

plt.show()
```
